```python
print("Document Summary: ")
print("----------------------------------------------------")
from nltk import sent_tokenize
sent_text = sent_tokenize(text)
num_sent = len(sent_text)
num_words = len(text.split())
print("Number of sentences:", num_sent)
print("Total number of words:", num_words)
count = {}
for word in text.split():
   if word in count :
      count[word] += 1
   else:
      count[word] = 1
num_uqe_words = len(count)
print("Number of Unique words:",num_uqe_words)
from gensim.summarization import summarize
print("----------------------------------------------------")
print("Summary: ")
print("                                                    ")
summari = summarize(text, split=True, ratio=0.1)
print(summari)
```
